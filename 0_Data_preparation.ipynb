## Function to obtain data set from zip link

```python
def zip2df(link):
    zip_path = tf.keras.utils.get_file(origin=link, fname=link.split('/')[-1], extract=True)
    csv_path = zip_path.replace('zip','csv')
    print(csv_path)
    df = pd.read_csv(csv_path)
    return df
```

## Crash data

```python
link_crash='https://github.com/duonghung86/Fatality-crashes/raw/master/Data/extract_2018_20191217230653_crash_20190601-20191130Texas.zip'
df_crash=zip2df(link_crash)
df_crash.tail()
```

## Time

```python
# Reorder crash date and time variable and convert it to datetime object
time=df['Crash_Date']+' '+df['Crash_Time']
time=pd.to_datetime(time)
```

```python
df.drop(columns=['Crash_Date','Crash_Time'],inplace=True)
```

```python
df['Crash_month']=time.apply(lambda x: x.month)
df['Crash_day']=time.apply(lambda x: x.day)
df['Crash_hour']=time.apply(lambda x: x.hour)
df['Crash_min']=time.apply(lambda x: x.minute)
df.head()
```

```python
# We need to convert Day of week to number
df['Day_of_Week'].value_counts()/len(df)*100
```

```python
map_dict = {'SAT':7, 'WED':4, 'TUE':3, 'SUN':1, 'MON':2, 'THU':5, 'FRI':6}
df['Day_of_Week'] = df['Day_of_Week'].map(map_dict)
```

```python
df['Day_of_Week'].isin([1,7])
```

```python
# Create a variable that determine crash date is in weekend or not
df['Weekend'] = np.where(df['Day_of_Week'].isin([1,7]),1,0)
df['Weekend'].value_counts()/len(df)*100
```

## Convert all possible float variables to integer

```python
for var in df.columns:
    if df[var].dtypes=='float':
        if var not in ['Latitude','Longitude']:
            print(var)
            df[var] = df[var].astype('int')
df.dtypes.unique()
```
