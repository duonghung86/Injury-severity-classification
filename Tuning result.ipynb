```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('VCA_Tuning1.csv')
df2 = pd.read_csv('VCA_Tune2.csv')
```

```python
df = pd.concat((df1,df2),ignore_index= True)
df
```

```python
index = df['Unnamed: 0'].str.split('-')
```

```python
df['type'] = [x[1] for x in index]
df['trial'] = [x[2] for x in index]
df.head()
```

```python
metric = ['Accuracy', 'AUC', 'G-mean', 'Avg_Pfm']
fig = plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    sns.boxplot(data=df,x=metric[i],y='type')
    
```

```python
df.groupby('type').mean().sort_values(by='AUC',ascending=False)[:3]
```

```python
df.groupby('type').mean().sort_values(by='G-mean',ascending=False)[:3]
```
